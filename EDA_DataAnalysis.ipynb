```python
formatted_df = pd.to_datetime(data["Application Date"])
data["Application Date"]=formatted_df.dt.strftime("%m%Y")
print(data['Application Date'])
```

```python
#duplicateID = data[data.duplicated(subset='Unique Record ID', keep=False)]
#print(duplicateID)

# making a bool series 
bool_series = data["Unique Record ID"].duplicated(keep=False) 
  
# displaying data 
data.head() 
  
# display data 
data[bool_series] 
```

```python
conditions = [
          (data['Worst Performance Everrrr'] == 0) | (data['Worst Performance Everrrr'] == 1),
          (data['Worst Performance Everrrr'] == 2) | (data['Worst Performance Everrrr'] == 3),
          (data['Worst Performance Everrrr'] == 4) | (data['Worst Performance Everrrr'] == 6),
          (data['Worst Performance Everrrr'] == 7),
          (data['Worst Performance Everrrr'] == 8)
           ]

# create a list of the values we want to assign for each condition
values = ['Good', 'Bad/Delinquent', 'Bad/Loss', 'Reject', 'Approved Not Booked (ANB)']

# create a new column and use np.select to assign values to it using our lists as arguments
data['Final Performance Flag'] = np.select(conditions, values)

# display updated DataFrame
data.head()
data.tail()
```

```python
# Get a bool series representing which row 
# satisfies the condition i.e. True for 
# row in which 'College' is 'Geu' 
data1 = data.apply(lambda x : True
            if x['Final Performance Flag'] == "Bad/Delinquent" or x['Final Performance Flag'] == "Bad/Loss" else False, axis = 1) 
  
# Count number of True in the series 
Bads = len(data1[data1 == True].index) 
  
print('Number of Bads: ', 
      Bads )
```

```python
data2 = data.apply(lambda y : True
            if y['Final Performance Flag'] == "Good" or y['Final Performance Flag'] == "Bad/Delinquent" or y['Final Performance Flag'] == "Bad/Loss" else False, axis = 1) 
  
# Count number of True in the series 
Booked = len(data2[data2 == True].index) 
  
print('Number of Booked: ', 
      Booked)
```

```python
Bad_Rate_Calculation= Bads/Booked
print(round(Bad_Rate_Calculation*100,2))
```
