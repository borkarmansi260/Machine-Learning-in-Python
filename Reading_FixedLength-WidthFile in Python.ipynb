```python
def col_interval_generator(start_position, field_length, has_binary = False):
    column_intervals = []
    for i in range(len(start_position)):
        if (not has_binary):
            column_interval = (start_position[i] - 1, start_position[i] - 1 + field_length[i])
            column_intervals.append(column_interval)
        else:
            if (i != (len(start_position) - 1)):
                column_interval = (start_position[i] - 1, start_position[i] - 1 + field_length[i])
            else:
                column_interval = (start_position[i] - 1, start_position[i] - 1 + field_length[i] * 2)
            column_intervals.append(column_interval)
    return column_intervals
```

```python
def read_data(filePath, fileType,firstLineHeader_T2, columnNames_T2, startPosition_T2, 
              fieldLength_T2, fieldConv_T2):
    if fileType == 2:
        column_intervals = col_interval_generator(start_position=startPosition_T2, field_length=fieldLength_T2, has_binary=False)
        if firstLineHeader_T2 == 1:
            df6 = pd.read_fwf(filePath, colspecs=column_intervals, header=0, converters=fieldConv_T2)
        else:
            df6 = pd.read_fwf(filePath, colspecs=column_intervals, header=None, names=columnNames_T2, converters=fieldConv_T2)
    return df6
```
