```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_excel('SampleData.xlsx')
print(data)
data.head()
data.tail()
```

```python
#Set Seed and generate numbers randomly- Random Number column
np.random.seed(10)
data['randNumCol'] = np.random.randint(1, 1000000, data.shape[0])
```

```python
print(data['randNumCol'])
```

```python
#Sort by the new column randNumCol
data = data.sort_values(by=['randNumCol'], ascending=True)
```

```python
#add a column for generating incremental sequence of numbers to use it for Unique_Id generation
data['Auto_SEQ'] = range(0, 0+len(data))
print(data['Auto_SEQ'])
```

```python
#convert the two columns to object datatype to combine them to form the unique ID

data['Auto_SEQ']=data['Auto_SEQ'].astype(str)
data['Unique Record ID']=data['Unique Record ID'].astype(str)
```

```python
data.dtypes
```

```python
data["Name Length"]=data["Unique Record ID"].str.len()
print(data["Name Length"])

#data.dtypes

if (np.any(data["Name Length"] != 6)):
    data["Unique Record ID"]=  data["Unique Record ID"].str.zfill(6)
else: 0
    
```
